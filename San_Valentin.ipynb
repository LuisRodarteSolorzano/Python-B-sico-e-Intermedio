# **Feliz día de San Valentin**

```python
# Aquí se importan las librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Aquí se construye el corazón
t = np.arange(0, 2*np.pi, 0.1)
xhrt = lambda t: 16*np.sin(t)**3
yhrt = lambda t: 13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)

dat = {'t': t, 'x': xhrt(t), 'y': yhrt(t)}
dat = pd.DataFrame(dat)

plt.figure(figsize=(15,10))
plt.plot(dat['x'], dat['y'], 'red')
plt.fill(dat['x'], dat['y'], 'magenta')
plt.text(-6.5, 0.55, 'I love you!', fontsize=60, color='blue')
plt.show()
```
